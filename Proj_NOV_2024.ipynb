```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Cargar los datos
file_path = 'D:/Escritorio/First period - 2024/Project - Econometrics/textfiles/HPRICE2.raw'
data = pd.read_csv(file_path, delim_whitespace=True)  # Ajustar según formato exacto de los datos

# Inspeccionar los datos
print(data.head())

# Verificar la Linealidad: Descripción básica de variables
print(data.describe())

# Verificar el muestreo aleatorio (no verificable directamente en datos)
# -> Esto depende del diseño de recolección de datos, lo asumimos en este caso.

# Verificar Multicolinealidad
# Calcular la matriz de correlación
correlation_matrix = data.corr()
print("Matriz de correlación:")
print(correlation_matrix)

# Calcular el Factor de Inflación de la Varianza (VIF)
def calculate_vif(df):
    vif_data = pd.DataFrame()
    vif_data["feature"] = df.columns
    vif_data["VIF"] = [
        sm.OLS(df[col], sm.add_constant(df.drop(columns=[col]))).fit().rsquared
        for col in df.columns
    ]
    return vif_data

# Eliminamos variables categóricas o no numéricas si hay
numerical_data = data.select_dtypes(include=[np.number])
vif = calculate_vif(numerical_data)
print("VIF para cada variable:")
print(vif)

# Verificar Exogeneidad: Test de correlación entre residuos y variables independientes
# Estimamos el modelo OLS
y = data['dependent_variable']  # Reemplazar por el nombre de la variable dependiente
X = data.drop(columns=['dependent_variable'])  # Reemplazar eliminando la dependiente
X = sm.add_constant(X)  # Agregar constante al modelo

model = sm.OLS(y, X).fit()
residuals = model.resid

# Correlación entre residuos e independientes
for col in X.columns[1:]:  # Excluyendo la constante
    correlation = np.corrcoef(residuals, X[col])[0, 1]
    print(f"Correlación entre residuos y {col}: {correlation}")

# Gráficos de diagnóstico
# Gráfico de residuos
plt.scatter(model.predict(X), residuals)
plt.axhline(0, color='red', linestyle='--')
plt.title("Residuos vs Valores Ajustados")
plt.xlabel("Valores Ajustados")
plt.ylabel("Residuos")
plt.show()

# Histograma de residuos
plt.hist(residuals, bins=30)
plt.title("Histograma de Residuos")
plt.xlabel("Residuos")
plt.ylabel("Frecuencia")
plt.show()
```
